```python
def mk_request(url):
    '''
    Takes the url and gives you status and returns responses
    '''
    response = requests.get(url)
    if 200 <= response.status_code < 400:
        return response
    else:
        print(f"request returned {response.status_code}error")
```

```python
## store the url

myurl = "https://collection.rubinmuseum.org/search/tibet"
```

```python
mk_request(myurl)
```

```python
## store the response
response = mk_request(myurl)
```

```python
##making soup
def mk_soup(response):
    '''
    makes soup out of the response you give
    '''
    return BeautifulSoup(response.text, "html.parser")
```

```python
## mother function that makes request and soup
def scrapes(url):
    '''
    give it a url to return docsoup of a page
    '''
    return mk_soup(mk_request(url))
```

```python
##scraping the url
tibet_soup = scrapes(myurl)
tibet_soup
```

```python
##narrowing the net down

all_objects = tibet_soup.find(id="timagesview")
all_objects
```

```python
##narrowing down the title

tibetan_objects = all_objects.find_all("div", class_="title")
tibetan_objects
```

```python
tibetan_links = []

for tibetan_object in tibetan_objects:
    tibetan_links.append(tibetan_object.find_all("a"))
```

```python
tibetan_links
```

```python
import itertools
```

```python
## there are lists within lists, we need to iter through
tibetan_links = list(itertools.chain(*tibetan_links))
```

```python
tibetan_links
```

```python
##get the hrefs
hrefs = [link.get("href") for link in tibetan_links]
hrefs
```

```python
base_url = "https://collection.rubinmuseum.org"
```

```python
for href in hrefs:
    full_urls = base_url + href
    print(full_urls)
```

```python
##creating snoozers

def snoozer(start_range, end_range):
    snooze_time = randrange(start_range, end_range)
    print(f"\n Snoozing for {snooze_time}seconds")
    return time.sleep(snooze_time)
```
