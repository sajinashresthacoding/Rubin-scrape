```python
def mk_request(url):
    '''
    Takes the url and gives you status and returns responses
    '''
    response = requests.get(url)
    if 200 <= response.status_code < 400:
        return response
    else:
        print(f"request returned {response.status_code}error")
```

```python
##making soup
def mk_soup(response):
    '''
    makes soup out of the response you give
    '''
    return BeautifulSoup(response.text, "html.parser")
```

```python
## mother function that makes request and soup
def scrapes(url):
    '''
    give it a url to return docsoup of a page
    '''
    return mk_soup(mk_request(url))
```

```python
## re-scrape this when you have time to append and even out the numbers
province = []
broken_url = []
object_numbers = []
donators = []
on_views = []
name = []
total_links = len(hrefs)
counter = 1

for href in hrefs:
        print(f"Scraping {counter} of {total_links}")
        counter+= 1
        try:
            code = scrapes(href)
            name_holder = code.find("h1", property="name")
            print(name_holder.get_text())
            name.append(name_holder.get_text())
            
            place_holder = code.find("div", class_="cultureField")
            culture_holder = place_holder.contents[1]
            print(culture_holder.get_text())
            province.append(culture_holder.get_text())
            
            donator_holder = code.find("div", class_="creditlineField")
            credit_holder = donator_holder.contents[1]
            print(credit_holder.get_text())
            donators.append(credit_holder.get_text())
            
            number_holder = code.find("div", class_="invnoField")
            number = number_holder.contents[1]
            print(number.get_text())
            object_numbers.append(number.get_text())
            
            view_holder = code.find("div", class_="onviewField")
            view = view_holder.contents[1]
            print(view.get_text())
            on_views.append(view.get_text())
     
        except:
            print(f"{href} was busted")
            broken_url.append(f"{href}")
            province.append(f"grab it yourself")
            donators.append(f"grab it yourself")
            object_numbers.append(f"grab it yourself")
            on_views.append(f"grab it yourself") 
            
        finally:
            snooze = randrange(10, 15)
            print(f"Snoozing for {snooze} seconds")
            time.sleep(snooze)
        print("*************************************************")

print("Done scraping all units")
```
