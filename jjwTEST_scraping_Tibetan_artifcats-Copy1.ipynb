```python
def mk_request(url):
    '''
    Takes the url and gives you status and returns responses
    '''
    response = requests.get(url)
    if 200 <= response.status_code < 400:
        return response
    else:
        print(f"request returned {response.status_code}error")
```

```python
##making soup
def mk_soup(response):
    '''
    makes soup out of the response you give
    '''
    return BeautifulSoup(response.text, "html.parser")
```

```python
## mother function that makes request and soup
def scrapes(url):
    '''
    give it a url to return docsoup of a page
    '''
    return mk_soup(mk_request(url))
```

```python
t_url = open("websites/tibet_search_origin.html", "r")
```

```python
t_soup=BeautifulSoup(t_url,"html.parser")
```

```python
##narrowing the net down

all_objects = t_soup.find(id="timagesview")
all_objects
```

```python
##narrowing down the title

tibetan_objects = all_objects.find_all("div", class_="title")
tibetan_objects
```

```python
##finding all the links that hold the access to the items

tibetan_links = []

for tibetan_object in tibetan_objects:
    tibetan_links.append(tibetan_object.find_all("a"))
```

```python
tibetan_links
```

```python
import itertools
```

```python
## there are lists within lists, we need to iter through
tibetan_links = list(itertools.chain(*tibetan_links))
```

```python
##get the hrefs
hrefs = [link.get("href") for link in tibetan_links]
hrefs
```
